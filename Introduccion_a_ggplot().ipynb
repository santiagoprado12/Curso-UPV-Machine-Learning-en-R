# DATOS+ESTETICA+CAPAS+FACETAS()

```
data(iris)
str(iris)
```

```
p<-ggplot(iris) #se crea un grafico vacio
p<-p+aes(x=Petal.Length, y=Petal.Width,colour=Species)#se agrega una estetica al grafico vacio
p
p<-p+geom_point()
p
p<-p+geom_smooth()
p
p<-p+facet_grid(~Species)
p
p<-p+ggtitle("Petal length and width")+labs(x="Petal length",y="petal width",colour="Species")
p
p<-p+theme_bw()
p
ggplot(iris,aes(x=Petal.Length,fill=factor(Species)))+geom_bar(stat='count')+facet_grid(.~Species)
```
